# RSI (RELATIVE STRENGTH INDEX)

1. As a momentum indicator, the relative strength index compares a security's strength on days when prices go up to its strength on days when prices go down

2. RSI= 100-[100/(1+(Average gain/Average Loss)]                                        ##following BODMAS strictly

3. The average gain or loss used in this calculation is the average percentage gain or loss during a look-back period.

4. Average %Gain/Loss= (%Gain/Loss)/no of days in look back period.

5. We can call this RS= Average Gain/Average Loss 
-RS= Relative Strength over a period
